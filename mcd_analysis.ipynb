Data Exploration, Analysis and Visualization on the McD Menu dataset

Let's have a look at the data by using the head() function. The dataframe.head() function displays top 5 records of the dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
pd.set_option('max_colwidth', None)
df=pd.read_csv(r'menu.csv')
df.head()
```

We can see that there are columns for category of food/beverage item, item name, portion/size of the item followed by values of various nutrients present in the item

Now let's take a look at the record counts per column and data type of each column

```python
df.info()
```

We see that there are no nulls in the dataset and there are 260 records in total

Data Stats -

Typical statistics of the dataset can be viewed using dataframe.describe()

```python
df.describe()
```

There is a plethora of insights that we can obtain from these stats but to keep it short, let me take an example of Cholesterol. 

If it is to be believed that Cholesterol should not take up more than 10% of our daily diet, we can analyse the column Cholesterol (% Daily Value) to see how McD items adhere to this requirement.

```python
df['Cholesterol (% Daily Value)'].plot(kind='kde')
```

The distribution and numerical stats convey that 50% of the items have Cholesterol (% Daily Value) < 11 which is a good provision for diet conscious consumers. This can further be strengthened by the fact that 25% of the items have Cholesterol (% Daily Value) < 2 which gives a good number of healthy options to choose from without limiting the entire day's cholesterol consumption to one McD item.

Now let us understand how visualization can help in identifying hidden facts in raw data.
To start with, we group the items based on their Category and calculate average value of the sugar content present in them.
Then, by plotting a horizontal bar graph, we can identify which category has the most sugar and what the numbers look like for other categories as well.

```python
df1=df.groupby(['Category']).mean()['Sugars']
df1.plot(kind='barh',figsize=(10,5))
```

Smoothies and shakes have the highest amount of sugar, and the average sugar content is double than the second highest category i.e. Coffee & tea. Below this, beverages and Desserts have similar numbers and are more than double than remaining categories like salads, breakfast, etc. These findings simply support the fact that beverages generally contain more sugar and should be consumed in moderation.

```python
df1=df.groupby(['Category']).mean()['Protein']
df1.plot(kind='bar',figsize=(15,5),grid=True)
```

From the bar chart above, we can infer that chicken and fish have the highest average protein content, followed by beef and pork. This is consistent with the fact that meat is protein-rich. The least amount of protein is present in beverages.

```python
df1=df.groupby(['Category']).mean()['Calories']
df1.plot(kind='line',figsize=(15,5),grid=True)
```

Chicken and fish have the highest amount of average calories, followed by breakfast and smoothies and shakes which have almost the same quantity of average calories. Since breakfast is an umbrella category, it would be interesting to dig deeper into it and identify what specific items make it so rich in calories.

```python
df1=df[df['Category']=='Breakfast'][['Item','Calories']]
df1.sort_values(by='Calories',ascending=False).head(10)
```

What is amusing to see here is that the amount of calories in a big breakfast with hotcakes reduces when egg whites are added to it. Thus, this is definitely something that consumers should consider while ordering; same goes for a normal big breakfast without hotcakes. Also, it can be observed that an egg and cheese bagel is more rich in calories if eaten with steak instead of bacon.

```python
df.groupby(['Category']).max()['Calcium (% Daily Value)'].sort_values(ascending=False)
```

Here, we can see that smoothies and shakes have the highest amount of Calcium % DV that an item can ever have, closely followed by coffee and tea. This makes sense because beverages (assuming dairy based) are known to be rich in calcium. 

```python
df.groupby(['Category']).count()['Item'].plot(kind='pie')
```

This is to identify how the distribution of items on the menu is. Clearly there are 40%(~104) items that belong to coffee and tea category and the least number of options are present for salads and desserts. McD being a fast food chain this is understandable. However, the high number of coffee and tea items intrigues me.

```python
df1=df[df['Category']=='Coffee & Tea'][['Item','Serving Size']]
df1['Item split']=df1['Item'].str.split(' \(').str[0]
print(df1.head())
print('Unique number of drinks is: ',df1['Item split'].nunique())
```

Thus, although there are ~104 items in the menu for coffee and tea category, there are only 31 different kinds of drinks which get multiplied by their serving sizes. I am sure this duplication holds true for other items as well, but I thought of investigating this for one category just to be sure.

```python
df['cal_from_fat_perc']=(df['Calories from Fat']/df['Calories'])*100
df['cal_from_fat_perc'].fillna(0.0,inplace=True)
df[['Item','cal_from_fat_perc']].sort_values(by='cal_from_fat_perc',ascending=False,ignore_index=True).head(10)
```

This calculation is to identify what kind of items have the most calories that are contributed by fat. Turns out that biscuits and nuggets occupy a large number of top spots here and should be consumed carefully.

```python
df1=df[df['Category']=='Snacks & Sides'][['Category','Serving Size','Item']]
df1['Ounces']=df1['Serving Size'].str.split(' oz').str[0]
df1[['Item','Ounces']].sort_values(by='Ounces',ascending=False)
```

Here we simply focuss on snacks and sides and compare portion sizes to determine weight-wise distribution of all items. We use string operations here as the ounces column values cannot be used as is. Large french fries are the heaviest of them all and we see that although there is a considerable difference between the weights of large and medium french fries, the small one is not that tiny as compared to the medium one. It is also amusing to know that different kinds of wraps with crispy chicken are just around 0.2 ounces heavier than their grilled versions - this is most probably because of the extra covering that goes into making chicken crispy.

```python
df1=df[['Item','Category','Carbohydrates','Protein','Total Fat']]
df1['Total Fat']=df1['Total Fat'].astype(int)
df1['Healthy']=(df1['Total Fat']>((df1['Carbohydrates']/2)-10)) & (df1['Total Fat']<((df1['Carbohydrates']/2)+10)) & (df1['Carbohydrates']>df1['Total Fat']) & (df1['Total Fat']>df1['Protein'])
df1.groupby(['Category','Healthy']).count()['Item']
```

Let's assume that an item is considered to be healthy if it follows a convention that the amount of total fat present in it falls between +10 and -10 of the half of amount of carbohydrates it contains, and the general trend is carbohydrates > total fat > protein.
After applying this to all items, we find out that only snacks and sides have more healthy options than unhealthy ones. All beverages, smoothies and shakes are unforgivably unhealthy.
